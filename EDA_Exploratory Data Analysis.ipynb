```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('data/young_survey.csv')
df.head()
```

```python
# 기본정보 가져오기
basic_info=df.iloc[:, 140:]
basic_info.head()
```

```python
# 기본정보에 대해서 통계적으로 파악하기
# 숫자만 통계 가능 , 문자는 describe()로하는 통계가 힘들다.
basic_info.describe()
```

```python
# 이렇게 문자는 직접 통계를 확인함
basic_info['Gender'].value_counts()
```

```python
basic_info['Handedness'].value_counts()
```

```python
basic_info['Education'].value_counts()
```

```python
# 그래프를 이용해 파악
sns.violinplot(data=basic_info, y='Age')
```

```python
# 성별에 따른 나이 분포
sns.violinplot(data=basic_info,x='Gender',y='Age')
```

```python
sns.violinplot(data=basic_info,x='Gender',y='Age',hue='Handedness')
```

```python
# jointplot 은 연관성을 보기 좋고
# 각각의 히스토그램도 그려져 있다.
sns.jointplot(data=basic_info, x='Height',y='Weight')
```

```python
# 상관관계를 보여주는 함수
df.corr()
```

```python
df.corr()['Age'].sort_values(ascending=False)
# 이거 보면 숫자로 나오는 숫자가 작을수록 관계가 없는거다.
# 1에 가까울수록 일차함수와 같은건데
# 1등이 weight인데 나이가 많아질수록 몸무게가 무거워 지는게 0.2정도 된다는거다.
# 숫자가 작아질 수록 반비례 관계라는 거다.
# 즉, 여기써있는 숫자가 일차함수의 기울기라고 생각하면 되겠다.
```

클러스터 분석  
뭉쳐있는 무리 -> 분석  
즉, 데이터들을 몇가지 부류로 나눠서 분석한다.  

```python
df.head()
```

```python
interests =df.loc[:,'History':'Pets']
interests.head()
```

```python
corr = interests.corr()
corr.head()
```

```python
corr['History'].sort_values(ascending=False).head(10)
```

```python
# 한 컬럼마다 이렇게 상관관계를 다 파악하기 어려우니까
# clustermap을 사용한다.
# 이건 연관성 있는 컬럼끼리 묶어져 있다.
sns.clustermap(corr)
```
