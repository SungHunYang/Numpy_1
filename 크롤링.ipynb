```python
# 처음 시작할 때
import requests # 얘는 '아자스'할때 얘만
from bs4 import BeautifulSoup # 얘는 샐래니움(가상 크롬) 하려면 얘가 있어야함.
```

```python
# 크롤링을 할 사이트 선택
targetSite = 'http://www.dowellcomputer.com/main.jsp'

# requests 모듈의 get() 메소드로 타겟 사이트의 정보를 얻어온다. 요정한다.
# requests.get(타겟 사이트의 주소)
# 지금 여기서는 타겟의 정보를 get이 얻어와서 request한테 넘겨준거다.
request = requests.get(targetSite)

# <Response [200]> (사이트 정보를 정상적으로 얻어왔다.) 나오면 크롤링이 가능한 사이트 
# <Response [405]> 나오면 크롤링 불가 사이트
print(request)

# request에 있는 정보중 text 글씨만 가져오겠다
# 이렇게 정보를 가져오는것까지가 request가 하는 일
html = request.text
print(html)
```

```python
# bs4 모듈의 BeautifulSoup() 함수로 타겟 사이트에서 얻어온 html 문서를 파싱할 수 있도록 준비한다.
# BeautifulSoup(이용할 사이트 정보,'html.parser') 
# 여기서 뒤에 html.parser는 html로 분석하겠다는건데, 이게 html이 아닐 수도 있다 우리가 아는 사이트가 html이 아니면 다른값을 적어야한다.
# 여기까지가 크롤링을 하기위한 준비
soup=BeautifulSoup(html,'html.parser')
print(soup)
```

```python
# html 태그와 html 태그에 지정한 속성(class)과 속성값을 이용한 크롤링
# findAll('태그이름',{'속성이름' : '속성값'}) => 태그의 속성이 지정되어 있지 않으면 {} 안의 내용은 생략할 수 있다.

# 최근 공지사항은 <b> 태그 내부에 있다.
# <b> 라는 태그만 있고 별도의 속성은 없었다.
notices = soup.findAll('b')

# print(notices)

# for notice in notices:
#    print(notice)
# 이렇게 되면 태그 <b> 까지 같이 나오는데
# 태그 없이 글씨만 보고 싶으면 notice.text 하면 된다.

for notice in notices:
    print(notice.text)
```

```python
# 최근 공지사항은 class 속성이 tail인 <td> 태그 내부에 있다.
# 아까 태그 <b>를 감싸고 있던 속성이다.
# 그니까 태그<b>로만 해서 찾았을 때는 겹쳐지는게 너무 많아서 원하는 값보다 더 많은 값이 나오니까
# 내가 원하는 값만 가지고 있는 좀 더 넓은 부분을 찾은 거다.
notices = soup.findAll('td',{'class' : 'tail'})

# 그런데 결과가 더 많이 나와서 다시 수정한다.
for notice in notices:
    print(notice.text)
    
# 지금 하고 있는게 크롤링의 시도작업 중인거다. 여러번 해보고 찾고 원하는 값만 나올 수 있게
```

```python
# 태그만 가지고는 원하는 것을 찾기 어려워서 CSS 를 사용해서 찾는다. 
# CSS 란 html로 문서를 만들면 그 문서를 꾸며주는 역할이다. ( 그리고 그런 문서에 동작을 주는게 java script다.)
# 즉 html => 사이트 구현 , CSS => 사이트 꾸미기 , java Script => 사이트 동작 구현

# CSS 선택자를 이용한 크롤링
# 자식 선택자(>) 또는 자손 선택자( ) [공백 한칸] 를 이용한 크롤링
# select('CSS선택자')

# 최근 공지사항은 href(하이퍼링크 ( 그 클릭하면 다른사이트로 넘어가지는거)) 속성값에 'notice' 문자열이 포함되어 있는 <a> 태그 내부에 있다.
# td 의 자식 a => ('td > a') // td의 자손 a =>('td a')
# 이게 자식과 자손은 들여쓰기 된 부분인데 
# 자식은 바로 한칸 들여쓰기한(아빠) 것이고, 자손은 한칸보다 더 크게 들여쓰기한거 (할아버지) 이런 느낌
notices = soup.select('td > a')
# print(notices)

for notice in notices:
    # print(notice)
    # get() 메소드로 인수로 지정한 속성값을 얻어온다.
    # print(notice.get('href'))
    # 'notice' 가 포함된 요소의 텍스트만 얻어온다.
    if notice.get('href').find('notice') >= 0:
        print(notice.text)
```
