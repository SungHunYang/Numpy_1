```python
iphone_df3 = pd.read_csv("data/iphone.csv",index_col=0)
```

```python
# row 값을 추가 하기
# 이렇게 원래 테이블에 없던 값(iPhone XR) 으로 넣어주면 추가된다.
# 맨 밑에 추가 된다.
iphone_df3.loc['iPhone XR'] =['2018-10-26',6.1,'3GB','iOS 12.0.1','Yes']
```

```python
iphone_df3
```

```python
# col 값을 추가하기
# 이렇게 새로운 col이름 으로 하고 Apple을 넣으면 전부 Apple이 들어간다.
iphone_df3['제조사'] ='Apple'
```

```python
iphone_df3
```

```python
# 값 삭제하기
# drop을 사용해서 row를 삭제하기
# 원하는 컬럼 값을 적고 axis = 'index'를 하면 된다,
# 여기서 axis='index'는 pandas에서 row를 index라고 부르기 때문에
# axis = 축이라는 뜻, 즉 row를 기준으로 삭제해라는거다.
# 그리고 inplace 는 기존의 데이터를 건드냐 마냐를 선택하는거다.
# 즉 한번만 바뀌는게 보여지냐(False) 아니면 영구적으로(True) 바꾸냐이다.
# csv 자체를 바꾸는게 아니라, 영구적으로 바뀐다는게 여기서는
# iphone_df3의 값이 아예 바뀐다는 것이다.
# False로 하면 iphone_df3값이 아예 바뀌는게 아니라
# 그냥 print하듯이 없어진 값이 한번만 보이게 되고 원상복구 된다.
# 그래서 False를 하면 다른 값에 넣어주면 저장 할 수 있다.
# 예를 들어 iphone_df4 = iphone_df3.drop('iPhone XR', axis ='index', inplace=False)
# 이렇게
# 이름.drop('row명',axis='index',inplace=False)
iphone_df3.drop('iPhone XR', axis ='index', inplace=False)
```

```python
iphone_df3
```

```python
iphone_df3.drop('iPhone XR', axis ='index', inplace=True)
```

```python
# col을 삭제하기
# drop을 사용해서 col를 삭제하기
# 원하는 컬럼 값을 적고 axis = 'columns'를 하면 된다,
# axis = 축이라는 뜻, 즉 col를 기준으로 삭제해라는거다.
# 나머지는 row삭제와 동일
# 이름.drop('col명',axis='columns',inplace=False)
iphone_df3.drop('제조사', axis ='columns', inplace=True)
```

```python
iphone_df3
```

```python
#여러줄을 한꺼번에 삭제
iphone_df3.drop(['iPhone 7','iPhone 8','iPhone X'], axis ='index', inplace=True)
```
