```python
df= pd.read_csv('data/parks.csv')
df.head()
```

```python
# 문자열 분리
# split => 띄어쓰기를 기준으로 단어들을 나눠서 리스트처럼 만들어줌
df['소재지도로명주소'].str.split()
```

```python
# n=1 넣으면 그러면 띄어쓰기를 전부 나누는게 아니라 첫번째 띄어쓰기만 나눈다.
df['소재지도로명주소'].str.split(n=1)
```

```python
# expand =True 를 넣으면 새로운 데이터 프레임이 된다.
adress = df['소재지도로명주소'].str.split(n=1,expand=True)
adress
```

```python
# 이런거 넣을때 알고 있겠지만 원래의 데이터 프레임과 넣을 데이터 프레임의 길이는 같아야 한다.
df['관할구역']=adress[0]
df
```

```python
df=pd.read_csv('data/laptops.csv')
df.head()
```

```python
brand_nation={
    'Dell' : 'U.S.',
    'Apple': 'U.S.',
    'Acer' : 'Taiwan',
    'HP' : 'U.S.',
    'Lenovo' : 'China',
    'Alienware' : 'U.S.',
    'Microsoft' : 'U.S.',
    'Asus' : 'Taiwan'
}
```

```python
# brand_nation 에서 왼쪽의 있는 값이 오른쪽 값으로 전부 대체 된다.
df['brand'].map(brand_nation)
```

```python
df['brand_nation'] = df['brand'].map(brand_nation)
```

```python
df
```

```python
# 카테고리별로 분석하는걸 편하게 해주는 역할
# groupby
nation_groups=df.groupby('brand_nation')
type(nation_groups)
```

```python
# 각 컬럼에 몇개의 값이 있는지
# 국가별로 비교하면서
nation_groups.count()
```

```python
# 최대값
nation_groups.max()
```

```python
# 각 국가별로 평균값 비교
nation_groups.mean()
```

```python
# 각 국가별로 가장 첫번째로 나온 값을 보여줌
nation_groups.first()
```

```python
# 각 국가별로 가장 마지막으로 나온 값을 보여줌
nation_groups.last()
```

```python
nation_groups.plot(kind='box',y='price')
```

```python
nation_groups.plot(kind='hist',y='price')
```

```python
price_df = pd.read_csv('data/vegetable_price.csv')
quantity_df = pd.read_csv('data/vegetable_quantity.csv')
```

```python
price_df
```

```python
quantity_df
```

```python
# 두 데이터를 합병하는건데
# pandas의 merge를 사용한다.
# merge.(1번째 데이터, 2번째 데이터 , on = '합병의 기준이 될 컬럼')
# 이렇게 하면 inner join이 디폴트이다.
pd.merge(price_df,quantity_df, on='Product')
```

```python
# 이렇게 how를 통해서 'left', 'right', 'full' join을 선택 할 수있다.
pd.merge(price_df,quantity_df, on='Product', how='left')
```
